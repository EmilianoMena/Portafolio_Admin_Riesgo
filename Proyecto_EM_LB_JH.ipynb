```python
import yfinance as yf
data = yf.download("GOOG INTC GM OMAB NRG K", start="2021-01-01", end="2022-12-31")
```

```python
data=pd.DataFrame(data)
data=data[["Adj Close","Volume"]]
```

```python
adjclose=data.iloc[:,0:6]
rends=adjclose.pct_change()
datos=pd.concat([data,rends],axis=1)
datos
```

```python
rend_medio=rends.iloc[0:252,:].mean()
rend_medio
```

```python
voldiaria=rends.iloc[0:252,:].std()
voldiaria
```

```python
año1=pd.DataFrame({"rendimiento anual":252*rend_medio,"Volatilidad esperada":np.sqrt(252)*voldiaria})
año1
```

```python
S=np.diag(año1.iloc[:,1])
```

```python
corr=rends.iloc[0:252,:].corr()
corr
```

```python
from scipy.optimize import minimize
```

```python
## Construcción de parámetros
# 1. Sigma: matriz de varianza-covarianza Sigma = S.dot(corr).dot(S)
S=np.diag(año1.iloc[:,1])
Sigma = S.dot(corr).dot(S).astype(float)
# 2. Eind: rendimientos esperados activos individuales
Eind = año1.iloc[:,0].values.astype(float)
```

```python
# Función objetivo
def menos_rs(w, Eind, rf, Sigma):
    ep = Eind.dot(w)
    sp = (w.T.dot(Sigma).dot(w))**0.5
    rs = (ep - rf) / sp
    return -rs
```

```python
# Función objetivo
def varianza(w, Sigma):
    return w.T.dot(Sigma).dot(w)
```

```python
# Número de activos
n = len(Eind)
# Dato inicial
w0 = np.ones(n) / n
# Cotas de las variables
bnds = ((0, 1),) * n
# Restricciones
cons = {"type": "eq", "fun": lambda w: w.sum() - 1}
```

```python
# Portafolio de mínima varianza
minvar = minimize(fun=varianza,
                  x0=w0,
                  args=(Sigma,),
                  bounds=bnds,
                  constraints=cons,
                  tol=1e-10)
```

```python
minvar
```

```python
rf=0.03
```

```python
#Rendimiento, volatilidad, ratio de sharpe y pesos para el portafolio de minima varianza
w_minvar = minvar.x
e_minvar = Eind.dot(w_minvar)
s_minvar = (w_minvar.T.dot(Sigma).dot(w_minvar))**0.5
rs_minvar = (e_minvar - rf) / s_minvar
e_minvar, s_minvar, rs_minvar,w_minvar
```

```python
# Función objetivo
def menos_rs(w, Eind, rf, Sigma):
    ep = Eind.dot(w)
    sp = (w.T.dot(Sigma).dot(w))**0.5
    rs = (ep - rf) / sp
    return -rs
```

```python
# Número de activos
n = len(Eind)
# Dato inicial
w0 = np.ones(n) / n
# Cotas de las variables
bnds = ((0, 1),) * n
# Restricciones
cons = {"type": "eq", "fun": lambda w: w.sum() - 1}
```

```python
# Portafolio EMV
emv = minimize(fun=menos_rs,
               x0=w0,
               args=(Eind, rf, Sigma),
               bounds=bnds,
               constraints=cons,
               tol=1e-10)
```

```python
emv
```

```python
#Rendimiento, volatilidad y ratio de sharpe para portafolio EMV
w_emv = emv.x
e_emv = Eind.dot(w_emv)
s_emv = (w_emv.T.dot(Sigma).dot(w_emv))**0.5
rs_emv = (e_emv - rf) / s_emv
e_emv, s_emv, rs_emv,w_emv
```

```python
num_acc=np.floor(np.array((w_minvar*1000000)/data.iloc[252,0:6]))
posiciones=1000000*w_minvar
```

```python
ponderaciones=pd.DataFrame(columns=["GM","GOOG","INTC","K","NRG","OMAB"])
ponderaciones.loc["Ponderaciones"]=[w_minvar[0]*100,w_minvar[1]*100,w_minvar[2]*100,w_minvar[3]*100,w_minvar[4]*100,w_minvar[5]*100]
ponderaciones.loc["Capital asignado"]=np.array(1000000*w_minvar)
ponderaciones.loc["Precio por accion"]=np.array(data.iloc[252,0:6])
ponderaciones.loc["Numero de acciones"]=np.floor(np.array((w_minvar*1000000)/data.iloc[252,0:6]))
ponderaciones.loc["Capital usado"]=np.array(num_acc*data.iloc[252,0:6])
ponderaciones.loc["Capital sobrante"]=np.array(posiciones-num_acc*data.iloc[252,0:6])
ponderaciones
```

```python
año2=data.iloc[251:,0:6]
rends2=año2.pct_change().fillna(0)
rends2
```

```python
año2.index=pd.to_datetime(año2.index)
```

```python
rend_port=pd.DataFrame(columns=["Rendimiento Portafolio 01-01-2022 - 31-12-2022"])
rend_port["Rendimiento Portafolio 01-01-2022 - 31-12-2022"]=np.array((rends2*w_minvar).sum(axis=1))
rend_port.set_index(año2.index,inplace=True)
rend_port
```

```python
evol_port=pd.DataFrame(columns=["Rendimiento porcentual","Evolucion del capital","Perdida/ganancia dia"])
evol_port["Rendimiento porcentual"]=rend_port.iloc[:,0]
rendimientos=rend_port.iloc[:,0]
capital_inicial=(ponderaciones.iloc[4,:]).sum()
capital=[]
for rendimiento in rendimientos:
    capital_actual = capital_inicial * (1 + rendimiento)
    capital.append(capital_actual)
    capital_inicial=capital_actual
evol_port["Evolucion del capital"]=capital
evol_port["Perdida/ganancia dia"]=evol_port.iloc[:,0]*evol_port.iloc[:,1]
#rendimiento sin contar los dolares que sobraron de cada compra de acciones
evol_port
```
